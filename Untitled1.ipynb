# matplotib

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x=np.linspace (0,5,11)
y=x**2












```

```python
x
```

```python
y
```

```python
plt.plot(x,y,'b')
plt.xlabel('x Axis Title here')
plt.ylabel('y Axis Title here')
plt.title('string title here')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'s--')
plt.subplot(1,2,2)
plt.plot(y,x,'g*--')
```

```python
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y,'.g')
axes.set_xlabel('Generic x lable')
axes.set_ylabel('Generic y lable')
axes.set_title('Generic title')
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.8,0.4,0.8,0.7])
axes2=fig.add_axes([0.4,0.6,0.8,0.3])
axes1.plot(x,y,'b')
axes1.set_xlabel('x Label')
axes1.set_ylabel('y Label')
axes1.set_title('Axes 1 Title')

axes2.plot(y,x,'.')
axes2.set_xlabel('x2 Label')
axes2.set_ylabel('y2 Label')
axes2.set_title('Axes 2 Title')
```

```python
fig,axes=plt.subplots(figsize=(4,4))
axes.plot(x,y,'g*-')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('sanju')
```

```python
fig.savefig('graphsample.png',dpi=150)
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='x**2')
ax.plot(x,x**3,label='x**3')
ax.legend()
```
